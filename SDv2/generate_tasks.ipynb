```python

jobs = {
    'English Springer': ['Springer Spaniel'], 
    'Garbage Truck': ['Waste Collection Vehicle'], 
    'French Horn': ['Brass Horn'], 
    'Church': ['Chapel'], 
    'Cassette Player': ['Tape Player'], 
}

# write to .sh file, replace {English Springer} with job

# devices = [0, 1, 2, 3]
devices = [0,1,2,3]

os.makedirs('train_scripts/netfive', exist_ok=True)

ckpt_path = 'checkpoints/v2-1_512-ema-pruned.ckpt'

# get all tasks 
all_tasks = []
it = 0
for job, targets in jobs.items():
    for target in targets:
        device = it % len(devices)
        all_tasks.append((job, target, job.replace(' ', ''), target.replace(' ', ''), device))
        it += 1

# Create a script file for each device
for device in devices:
    file_name = f'train_scripts/netfive/erase_device_{device}.sh'
    with open(file_name, 'w') as f:
        f.write(f'# Tasks for device {device}\n\n')
        for it, task in enumerate(all_tasks):
            job, target, job_name, target_name, it_device = task
            if it_device == device:
                model_name = f'compvis-word_{job_name}-target_{target_name}-method_xattn-sg_3-ng_1-iter_1000-lr_1e-05-info_SDv21'
                f.write(f'# "Erasing {job_name} with target {target_name}"\n')
                f.write(f'CUDA_VISIBLE_DEVICES={device} python3 train-esd.py --seperator "," --train_method "xattn" --ckpt_path "{ckpt_path}" --diffusers_config_path "../Better_Erasing/models/erase/config.json" --prompt "{job}" --target "{target}" --config_path "configs/stable-diffusion/v2-inference-fp32.yaml" --info "SDv21" | tee "logs/erase_{job_name}_target_{target_name}_train.log"\n')
                f.write(f'CUDA_VISIBLE_DEVICES={device} python3 generate_images_ldm_sdv2.py --config="configs/stable-diffusion/v2-inference.yaml" --ckpt "models/{model_name}/{model_name}.pt" --prompts_path "data/netfive_100.csv" --H 512 --W 512 --n_samples 1 --outdir "evaluation_netfive_100/{model_name}/netfive_100" | tee -a "logs/erase_{job_name}_target_{target_name}_generate.log"\n')
                f.write(f'CUDA_VISIBLE_DEVICES={device} python3 eval-scripts/imageclassify.py --topk=10 --folder_path="evaluation_netfive_100/{model_name}/netfive_100/samples/" --prompts_path="data/netfive_100.csv" --save_path="evaluation_folder/{model_name}-netfive_100.csv"\n\n')
                f.write('\n')

```
